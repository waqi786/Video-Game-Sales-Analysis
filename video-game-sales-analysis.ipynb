<a id="A"></a>
# <div style="text-align:center; border-radius:15px 30px; padding:10px; color:white; margin:0; font-size:90%; font-family:Pacifico; background-color:red; overflow:hidden"><b> Handling Missing Values in 'Year' and 'Publisher' Columns </b></div>



```python
df["Year"] = df["Year"].fillna(df["Year"].median())
df["Publisher"] = df["Publisher"].fillna(df["Publisher"].mode()[0])
```

<a id="A"></a>
# <div style="text-align:center; border-radius:15px 30px; padding:10px; color:white; margin:0; font-size:100%; font-family:Pacifico; background-color:red; overflow:hidden"><b> Generating a Correlation Heatmap </b></div>



```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True, cmap="viridis", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

<a id="A"></a>
# <div style="text-align:center; border-radius:15px 30px; padding:10px; color:white; margin:0; font-size:100%; font-family:Pacifico; background-color:red; overflow:hidden"><b> Global Sales Trend Over the Years </b></div>



```python
df.groupby("Year")["Global_Sales"].sum().plot(kind="line", figsize=(12, 5), marker="o", color="red")
plt.title("Global Sales Trend Over the Years")
plt.xlabel("Year")
plt.ylabel("Total Global Sales (in millions)")
plt.grid(True)
plt.show()
```

<a id="A"></a>
# <div style="text-align:center; border-radius:15px 30px; padding:10px; color:white; margin:0; font-size:100%; font-family:Pacifico; background-color:red; overflow:hidden"><b> Top 10 Most Popular Gaming Platforms </b></div>



```python
plt.figure(figsize=(12, 5))
sns.barplot(x=df["Platform"].value_counts().index[:10], 
            y=df["Platform"].value_counts().values[:10], palette="bright")
plt.title("Top 10 Most Popular Gaming Platforms")
plt.xlabel("Platform")
plt.ylabel("Number of Games Released")
plt.xticks(rotation=45)
plt.show()
```

<a id="A"></a>
# <div style="text-align:center; border-radius:15px 30px; padding:10px; color:white; margin:0; font-size:100%; font-family:Pacifico; background-color:red; overflow:hidden"><b> Game Genre Distribution (Pie Chart) </b></div>



```python
df["Genre"].value_counts().plot(kind="pie", autopct="%1.1f%%", figsize=(8, 8), cmap="coolwarm")
plt.title("Game Genre Distribution")
plt.ylabel("")
plt.show()
```

<a id="A"></a>
# <div style="text-align:center; border-radius:15px 30px; padding:10px; color:white; margin:0; font-size:100%; font-family:Pacifico; background-color:red; overflow:hidden"><b> Top 10 Game Publishers by Number of Games Released </b></div>



```python
top_publishers = df["Publisher"].value_counts().nlargest(10)

plt.figure(figsize=(12,5))
sns.barplot(x=top_publishers.index, y=top_publishers.values, palette="plasma")
plt.title("Top 10 Game Publishers")
plt.xlabel("Publisher")
plt.ylabel("Number of Games")
plt.xticks(rotation=45)
plt.show()
```

<a id="A"></a>
# <div style="text-align:center; border-radius:15px 30px; padding:10px; color:white; margin:0; font-size:100%; font-family:Pacifico; background-color:red; overflow:hidden"><b> Top 10 Gaming Platforms by Global Sales </b></div>



```python
top_platforms = df.groupby("Platform")["Global_Sales"].sum().sort_values(ascending=False)[:10]
plt.figure(figsize=(12, 6))
sns.barplot(y=top_platforms.index, x=top_platforms.values, palette="magma")
plt.title("Top 10 Platforms by Global Sales")
plt.xlabel("Total Global Sales (in millions)")
plt.ylabel("Platform")
plt.show()
```

<a id="A"></a>
# <div style="text-align:center; border-radius:15px 30px; padding:10px; color:white; margin:0; font-size:100%; font-family:Pacifico; background-color:red; overflow:hidden"><b> Total Sales Distribution Across Regions </b></div>



```python
sales_data = df[["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales"]].sum()

plt.figure(figsize=(10, 5))
sns.barplot(x=sales_data.index, y=sales_data.values, palette="bright")
plt.title("Regional Sales Distribution")
plt.xlabel("Region")
plt.ylabel("Total Sales (in millions)")
plt.grid(True)
plt.show()
```
